```python
#Áª¥Â∫¶ÂèòÂåñÔºåÈúÄÈáçÊñ∞ÁîüÊàê
Sub_array = pd.to_numeric(data['Subscribers'])
View_array = pd.to_numeric(data['Video views'])
```

```python
div = Sub_array/View_array
```

```python
data['Division'] = div
```

```python
features.append('Division')
```
